This notebook demonstrates how to load and validate in vitro pharmacology and PK data for dose projection.

```python
import sys
sys.path.insert(0, '..')

from doseprojection.io import load_invitro_data, load_pk_data
import pandas as pd
```

## Load In Vitro Data

```python
invitro = load_invitro_data('../examples/example_invitro_data.csv')
invitro
```

## Load PK Data

```python
pk = load_pk_data('../examples/example_pk_data.csv')
pk
```

## Inspect Data Quality

Check for missing values in key columns.

```python
print("In vitro data — missing values:")
print(invitro.isnull().sum())
print()
print("PK data — missing values:")
print(pk.isnull().sum())
```

## Merge In Vitro + PK Data

Join on `compound_id` for integrated analysis.

```python
# Get rat IV PK data for dose projection
rat_iv = pk[(pk['species'] == 'rat') & (pk['route'] == 'IV')]
rat_po = pk[(pk['species'] == 'rat') & (pk['route'] == 'PO')]

# Merge with in vitro data
merged = invitro.merge(
    rat_iv[['compound_id', 'CL_mL_min_kg', 'Vss_L_kg']],
    on='compound_id', how='left'
)
merged = merged.merge(
    rat_po[['compound_id', 'F_pct']].rename(columns={'F_pct': 'F_pct_rat'}),
    on='compound_id', how='left'
)
merged[['compound_id', 'IC50_nM', 'fu_plasma_rat', 'CL_mL_min_kg', 'Vss_L_kg', 'F_pct_rat']]
```
